```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
import numpy as np
```

```python
def handle_outliers(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_indices = (column < lower_bound) | (column > upper_bound)
    column[outliers_indices] = np.nan  # Replace outliers with NaN
    return column
```

```python
df['Age'] = handle_outliers(df['Age'])

df['Fare'] = handle_outliers(df['Fare'])

df['SibSp'] = handle_outliers(df['SibSp'])
```

```python
df.fillna(df.mean(), inplace=True)
```
